```python
import pygame
import random
import os
white=(255,255,255)
red=(255,0,0)
black=(0,0,0)


pygame.init()
s_width=1300
s_height=700
gamewindow=pygame.display.set_mode((s_width,s_height))
pygame.display.set_caption("my first game")


clock=pygame.time.Clock()
def plot_snak(gamewindoe,color,snk_list,s_s):
    for x,y in snk_list:
        pygame.draw.rect(gamewindow,color,pygame.Rect(x,y,s_s,s_s))
        

font=pygame.font.SysFont(None,45)
def textscreen(text,color,x,y):
    screen_text=font.render(text,True,color)
    gamewindow.blit(screen_text,(x,y))
def welcome():
    exit_game=False
    while not exit_game:
        gamewindow.fill((250,240,220))
        textscreen("welcome to snake game", black ,510,300)
        textscreen("press space to play",black,530,330)
        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                exit_game=True
            if event.type==pygame.KEYDOWN:
                if event.key==pygame.K_SPACE:
                    gameloop()
        pygame.display.update()
        clock.tick(60)
def gameloop():
    exit_game=False
    game_over=False
    v_x=0
    v_y=0
    s_x=50
    s_y=40
    s_s=25
    f_x=random.randint(10,(s_width/4))
    f_y=random.randint(10,(s_height/4))
    fps=25
    score=0
    snk_list=[]
    snk_length=1;
    if(not os.path)
    with open("highscore.txt","r") as f:
        highscore=f.read()
    while not exit_game:
        if game_over:
            with open("highscore.txt","w") as f:
                f.write(str(highscore))
            gamewindow.fill(white)
            textscreen("game over!press enter to continue.",red,s_width/3,s_height/3)
            
            for event in pygame.event.get():
        #         print(event)
                if event.type==pygame.QUIT:
                    exit_game=True
                    
                if event.type==pygame.KEYDOWN:
                    if event.key==pygame.K_RETURN:
                        gameloop()
            
        else:
            for event in pygame.event.get():
        #         print(event)
                if event.type==pygame.QUIT:
                    exit_game=True
                if event.type==pygame.KEYDOWN:
                    if event.key==pygame.K_RIGHT:
                        v_x=10
                        v_y=0
                    if event.key==pygame.K_LEFT:
                        v_x=-10
                        v_y=0

                    if event.key==pygame.K_UP:
                        v_y=-10
                        v_x=0

                    if event.key==pygame.K_DOWN:
                        v_y=10
                        v_x=0
                    if event.key==pygame.K_SPACE:
                        score+=10
            s_x=s_x+v_x
            s_y=s_y+v_y

            if abs(s_x-f_x)<10 and abs(s_y-f_y)<10:
                score+=10
                f_x=random.randint(0,s_width)
                f_y=random.randint(0,s_height)
                snk_length+=3
                if score>int(highscore):
                    highscore=score

            gamewindow.fill((150,200,250))
            textscreen("Score: "+str(score)+" highscore: " + str(highscore),red,5,5)
            pygame.draw.rect(gamewindow,red,pygame.Rect(f_x,f_y,s_s,s_s))

            head=[]
            head.append(s_x)
            head.append(s_y)
            snk_list.append(head)

            if len(snk_list)>snk_length:
                del snk_list[0]
                
            if head in snk_list[:-1]:
                game_over=True

                
            if s_x<0 or s_x>s_width or s_y<0 or s_y>s_height:
                game_over=True
              

            plot_snak(gamewindow,black,snk_list,s_s)
        pygame.display.update()
        clock.tick(fps)
    pygame.quit()
    quit()
welcome()
```
